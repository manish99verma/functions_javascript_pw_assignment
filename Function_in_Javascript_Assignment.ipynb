2. Create a JavaScript function called generateGreeting that takes a name as an argument and returns a personalized greeting message. Use this function to greet three different people.

- Javascript code for it:
```
const generateGreeting = (name) => `Hello, ${name}!`;
const users = ["Alice", "Bob", "Charlie"];
for (let i = 0; i < users.length; i++) {
  console.log(generateGreeting(users[i]));
}
```

3. Create an IIFE (Immediately Invoked Function Expression) that calculates the square of a number and immediately displays the result.

- Javascript code for it:
```
(function (x) {
  const square = x * x;
  console.log(`The square of ${x} is ${square}`);
})(15);
```

4. Write a JavaScript function called calculateTax that takes an income as an argument and returns the amount of tax to be paid. Use a closure to handle different tax rates based on income ranges. Test the function with various incomes.
- Code given below:
```
const calculateTax = (annualIncome) => {
  if (annualIncome <= 500000) {
    return 0; // No tax for income up to 500,000
  } else if (annualIncome <= 1000000) {
    return Math.floor(annualIncome * 0.1); // 10% tax for income between 500,001 and 1,000,000
  } else if (annualIncome <= 2000000) {
    return Math.floor(annualIncome * 0.2); // 20% tax for income between 1,000,001 and 2,000,000
  } else {
    return Math.floor(annualIncome * 0.3); // 30% tax for income above 2,000,000
  }
};
const testCases = [
  100000, 400000, 800000, 1200000, 1800000, 2100000, 2500000, 3500000, 5000000,
];
for (let i = 0; i < testCases.length; i++) {
  const income = testCases[i];
  const tax = calculateTax(income);
  console.log(`Annual Income: ${income}, Tax: ${tax}`);
}
```

5.  Write a JavaScript function called factorial that calculates the factorial of a non-negative integer using recursion. Test the function with different inputs.
- Given program have `factorial` funtion which calculates factorial of non-negative integers recurcively with test cases that tests numbers from 1 to 20.
```
const factorial = (n) => {
  if (n === 1) return 1;
  return n * factorial(n - 1);
};
// Test
for (let i = 1; i < 20; i++) {
  console.log(`Factorial of ${i} is ${factorial(i)}`);
}
```

6.  Write a JavaScript function called curry that takes a function as an argument and returns a curried version of that function. The curried function should accept arguments one at a time and return a new function until all arguments are provided. Then, it should execute the original function with all arguments. Test the curry function with a function that adds two numbers.
- Curry function to add two numbers:
```
const add = (a) => {
  return function (b) {
    return a + b;
  };
};
console.log(`Result: `, add(6)(8));
```
